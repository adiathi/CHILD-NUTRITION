```python
#calculator code 
import tkinter as tk 
import math 
from functools import partial


#creating funtion for sum
def summ(n1,n2):
    result= n1+n2
    return result


# creating function for subtraction
def subt(n1,n2):
    result= n1-n2
    return result

#creting function for multiplication
def mult(n1,n2):
    result= n1*n2
    return result

#creting function for division
def division(n1,n2):
    if(n2==0):
        result= float('inf')
    else:
        result= n1/n2
    return result

#creting function for LCM
def lcm(n1,n2):
    result= math.lcm(n1,n2)
    return result


#creting function for HCF
def Hcf(n1,n2):
    result=0
    if n1 > n2:
        smaller = n2
    else:
        smaller = n1
    for i in range(1, smaller+1):
        if((n1 % i == 0) and (n2 % i == 0)):
            result = i 
    return result

#creting function for log
def log(n1,n2):
    if n2==0 or n2==1 :
            result=float('inf')
    else:
        result=math.log(n1,n2)
    return result

#creting function for power
def power(n1,n2):
    result= n1**n2
    return result



def call_result():
    n1 = int(number1.get())
    n2 = int(number2.get())
    selection = int(number3.get())
#IF ELIF stactement to find the operation
#OPERATION ADDITION
    if selection==1:
        tk.Label(root, text = "                                                         ").grid(row=10,column=300)
        result = summ(n1,n2)
        tk.Label(root, text ="Result = %d" %  result).grid(row=10,column=300)
        
#OPEARTION SUBTRACTION        
    elif selection==2:
        tk.Label(root, text = "                                                           ").grid(row=10,column=300)
        result=subt(n1,n2)
        
        tk.Label(root, text ="Result = %d" %  result).grid(row=10,column=300)
        
#OPERATION MULTIPLICATION
    elif selection==3:
        tk.Label(root, text = "                                                                       ").grid(row=10,column=300)
        result=mult(n1,n2)
        
        tk.Label(root, text ="Result = %d" %  result).grid(row=10,column=300)
        
#OPERATION DIVISION
    elif selection==4:
        tk.Label(root, text = "                                                                      ").grid(row=10,column=300)
        result=division(n1,n2)
        tk.Label(root, text ="Result = %d" %  round(result,8)).grid(row=10,column=300)
        
        
#OPERATION HCF
    elif selection==6:
        tk.Label(root, text = "                                                                       ").grid(row=10,column=300)
        result =Hcf(n1,n2)
                
        tk.Label(root, text ="Result = %d" %  round(result,8)).grid(row=10,column=300)
       
# Operation LCM
    elif selection==5:
        tk.Label(root, text = "                                                                       ").grid(row=10,column=300)
        result=lcm(n1,n2)
        
        tk.Label(root, text ="Result = %d" %  round(result,8)).grid(row=10,column=300)
        
#OPERATION EXPONENT
    elif selection==8:
        tk.Label(root, text = "                                                                        ").grid(row=10,column=300)
        result= power(n1,n2)
        
        tk.Label(root, text ="Result = %d" %  round(result,8)).grid(row=10,column=300)
        
#OPERATION LOGARITHM
    elif selection==7:
        tk.Label(root, text = "                                                                       ").grid(row=10,column=300)
        result=log(n1,n2)
        tk.Label(root, text ="Result = %d" %  round(result,8)).grid(row=10,column=300)
     
#INVALID OPERAION
    else:
        tk.Label(root, text = "                                                                        ").grid(row=10,column=300)
        tk.Label(root, text = "INVALID OPERATION").grid(row=10,column=300)

#Main snippite
root = tk.Tk() 
root.configure(background="#cceeff")
root.geometry('400x200+100+200')
root.title('Calculator')
number1 = tk. StringVar()
number2 = tk. StringVar()
number3 = tk. StringVar()
label1=tk.Label(root, text="CALCULATOR!", fg="#99004d").grid(row=0,column=300)
labelNum2 = tk.Label(root, text="Please select operation :-\n1. Add\n2. Subtract\n3. Multiply\n4. Division\n5. LCM\n6. HCF\n7. LOG\n8. Power" ).grid(row=1, column=0)
labelNum = tk.Label(root, text="Number1").grid(row=5, column=2)
labelNum7 = tk.Label(root, text="Number2").grid(row=6, column=2)
labelNum8 = tk.Label(root, text="Operation").grid(row=7, column=2)

entryNum1 = tk.Entry(root, textvariable=number1).grid(row=5, column=4)
entryNum2 = tk.Entry(root, textvariable=number2).grid(row=6, column=4)
entryNum3 = tk.Entry(root, textvariable=number3).grid(row=7, column=4)         

buttonCal = tk.Button(root, text="Calculate", fg='black', bg='#ffe6e6', command=call_result).grid(row=8, column=0)
root.mainloop()
```
